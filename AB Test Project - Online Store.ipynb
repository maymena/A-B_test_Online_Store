## Intro

**Task statement**

You've received an analytical task from an international online store. Your predecessor failed to complete it: they launched an A/B test and then quit (to start a watermelon farm in Brazil). They left only the technical specifications and the test results.

**Technical description**
- Test name: recommender_system_test
- Groups: А (control), B (new payment funnel)
- Launch date: 2020-12-07
- The date when they stopped taking up new users: 2020-12-21
- End date: 2021-01-01
- Audience: 15% of the new users from the EU region
- Purpose of the test: testing changes related to the introduction of an improved recommendation system
- Expected result: within 14 days of signing up, users will show better conversion into product page views (the product_page event), product card views (product_card) and purchases (purchase). At each of the stage of the funnel product_page → product_card → purchase, there will be at least a 10% increase.
- Expected number of test participants: 6000

Download the test data, see whether it was carried out correctly, study conversion at different funnel stages and analyze the results.

```python
def explore_initial_data(df_list):
    for df in df_list:
        print('\nTable:',df.name)
    # printing the general/summary information about the DataFrame 
        df.info()
    # Checking for other reasons and patterns that could lead to missing values  
        print('\n',df.name,'descrieptive:\n',df.describe(include='all'))

    # printing a sample of data
        print('\n',df.name,'sample:\n',df.head(15))
    # Checking unique values in the whole dataset
        columns=list(df)
# another way with faster runtime: columns=data.columns.values.tolist() 
        print('\nUnique values in each column of the table:',df.name)
        for column in columns:
            print('\nTable:',df.name,'Column:\n', column,'\n',df[column].value_counts(dropna=False).sort_index())
    
    # Checking duplicated rows
        print('\nDuplicated rows:\n')
        print(df.duplicated().sum())
    
    # Checking missing values
        print(df.stb.missing())
        display(df.stb.missing(clip_0=True, style=True))
        

# applying function
df_list=[marketing_events,users,user_events,participants]
explore_initial_data(df_list)
```

```python
# number of users that have been added to the `recommender_system_test` by each day
participants_day=users.loc[(users.user_id.isin(participants_recommender_system_test.user_id)),
                           ['user_id','first_date']].groupby('first_date').count().reset_index()

participants_day.columns=['first_date','n_users']

#print('Number of users that have been added to the `recommender_system_test` by each day:\n',participants_day)

# number of new users in the EU by each day
users_eu_day=users.loc[users['region']=='EU',['user_id','first_date']].groupby('first_date').count().reset_index()
users_eu_day.columns=['first_date','n_users']
#print("")
#print('Number of new users in the EU by each day:\n',users_eu_day)


# dropping days that are not relevant for the A/B test
eu=users_eu_day[~users_eu_day['first_date'].isin(['2020-12-22','2020-12-23'])]

# Percentages of participants in the `recommender_system_test` within EU new users by each day
participants_day['%within_new_EU']=(participants_day.n_users/eu.n_users)*100
print("")
print('Number of users that have been added to the `recommender_system_test` by each day and percentages of participants in the `recommender_system_test` within EU new users by each day :\n')
participants_day

# number of users that have been added to the `recommender_system_test` by each day by group
participants_day_group=users.loc[(users.user_id.isin(participants_recommender_system_test.user_id)),
                                 ['user_id','first_date']].merge(participants_recommender_system_test[['user_id', 'group']],
                                                                 on='user_id').groupby(['first_date','group'])\
                                                                 .user_id.count().reset_index()

participants_day_group.columns=['first_date','group','n_users']
#print("")
#print('Number of users that have been added to the `recommender_system_test` by each day by group :\n',participants_day_group)

# seperating groups
participants_day_group_a=participants_day_group.loc[participants_day_group['group']=='A'].reset_index()
participants_day_group_b=participants_day_group.loc[participants_day_group['group']=='B'].reset_index()



# claculating percentages within new EU users for each group
participants_day_group_a['%within_new_EU']=(participants_day_group_a.n_users/eu.n_users)*100

participants_day_group_b['%within_new_EU']=(participants_day_group_b.n_users/eu.n_users)*100

print("")
print("Group A: Number of users that have been added to the `recommender_system_test` by each \
day and percentages within new EU users:\n")

participants_day_group_a
print("")
print('Group B: Number of users that have been added to the `recommender_system_test` by each\
day and percentages within new EU users:\n')

participants_day_group_b
```

### A/B test analysis

#### Number of users in each group

```python
len(participants_recommender_system_test.query('group=="A"'))
len(participants_recommender_system_test.query('group=="B"'))
```

```python
# H0: proportions of group A and B are equale (each 50%)
# H1: proportions of group A and B are not equale

from scipy.stats import chisquare

# one sample chi-square test 
#a.k.a  goodness-of-fit test
frequencies = [len(participants_recommender_system_test.query('group=="A"')),len(participants_recommender_system_test.query('group=="B"'))] 
# by default, the expected frequencies is Uniformly distributed so there is no need for the expected proportions parameter: 
# result = chisquare(frequencies, [50, 50])
result = chisquare(frequencies)

print("pvalue =", '{:.4f}'.format(result.pvalue))
print("chisq =", '{:,.4f}'.format(result.statistic))

if result.pvalue < 0.01:
    print('Warning. Sample Ratio Mismatch may be present.')
else:
    print('Probably no SRM.')
```

#### Mistakes in the AB test

There is no point in preforming an independent z test of proportions to detect differences in conversions proportions for several reasons: 

- First, the results of the Z test will be biased:
  - As seen above, there is a significant difference between the sizes of groups A and B. 
  - Furthermore, the sample size are too small to detect a relative change of 10% at a 95% confidence level as expected in the technical description, except for the product page stage.
      - in the purchase stage the Baseline conversion rate is 32%  so the sample size should be 3,360 per each group.
      - in the product cart stage the Baseline conversion rate is 30%  so the sample size should be 3,692 per each group.
      - in the product page stage the Baseline conversion rate is 65%  so the sample size should be 854 per each group. So group A is larger but group B is smaller.  
      
        Calculations done with: https://www.evanmiller.org/ab-testing/sample-size.html (calculations with https://www.abtasty.com/sample-size-calculator/ are similar)
  - Moreover, the users were assigned to the control and test groups in an inconsistent manner (different percentage and quantities of users allocated to each group on each day) so the groups may have different types of customers.  This may compromise the validity of the results.

- Second, when looking at the funnels, Group B, the test group, preformed worse in all the stages. Even when looking at devices sub groups. So it doesn't matter if the results in this case are significant because even if they are (and they can be only at the product page stage) we still will not have a reason to set the new recommender system to all users.     

Hence, the A/B test should be preformed again, since it's current results will not be valid nor reliable. 

```python
funnel_by_groups
funnel_by_groups.loc[(funnel_by_groups['group']=='A') & (funnel_by_groups['event_name']=='purchase'),'n_users'].iloc[0]
```

```python
# H0: The proportions of the users at the event are the same in both groups

# H1: The proportions of the users at the event differs


from statsmodels.stats.proportion import proportions_ztest

# creating the function
def check_hypothesis(group_a,group_b,event,alpha=0.05):
    success_a=funnel_by_groups.loc[(funnel_by_groups['group']=='A') & (funnel_by_groups['event_name']==event),'n_users'].iloc[0]
    success_b=funnel_by_groups.loc[(funnel_by_groups['group']=='B') & (funnel_by_groups['event_name']==event),'n_users'].iloc[0]
    
    trials_a=len(participants_recommender_system_test.query('group=="A"'))
    trials_b=len(participants_recommender_system_test.query('group=="B"'))
    
    stat, pval = proportions_ztest([success_a,success_b], [trials_a,trials_b])
    print('{0:0.3f}'.format(pval))
    
    if (pval < alpha):
        print("Rejecting the null hypothesis for", event,"and groups", group_a,"and",group_b)
        print("This means the difference between the proportions of the samples is statistically significant")
    else:
        print("Failed to reject the null hypothesis for",event,"and groups", group_a,"and",group_b)
        print("This means the difference between the proportions of the samples is not statistically significant")

# applying the function
for i in sequence[1:]:
    check_hypothesis('A','B',i,alpha=0.05)
```

As seen above, several two proportions z-test were preformed. The difference between the proportions of the events in A/B groups is not statistically significant at the purcase and at the product cart stages. 

However, the difference between the proportions of the samples is statistically significant at the product_page stage.
 
The results remain the same even after a Bonferoni Correction (significance level of  0.05 /3 = 0.017)

Problems with the recommender A/B test:
  - There is a significant difference between the sizes of groups A and B. 
  - Furthermore, the sample size are too small to detect a relative change of 10% at a 95% confidence level as expected in the technical description (maybe except for the product page stage). 
  - Moreover, the users were assigned to the control and test groups in an inconsistent manner (different percentage and quantities of users allocated to each group on each day) so the groups may have different types of customers due to cyclical effects. This may compromise the validity of the results.
  - The uneven assigning was even more harmful due to the fact that in part of the A/B test period was a promotion campaign
  - Some of the participants participated in another A/b test at the same time.
  - There was a technical issue in the creation of time stamps of the events in each session.
  - not all users had data for 14 days from registration as the technical description required.  

**Recommendations**
Hence, a new A/B test should be preformed, since it's current results will not be valid nor reliable.

In the new A/B test: 
- The sample size should be at least 3,360 per each group and when tested the group sizes should be similar. 
- The assigning of participants should continue for 20 days and use 15% of new EU users.
- data about the last participant should be collected until 14 days from registration.
- Hence the experiment will continue for 34 days.
- A similar number of participants should be assigned in each day to each group
- Participants should be exclusively in one A/B test and in only one group.
- Different regions should be tested in other A/B tests.
